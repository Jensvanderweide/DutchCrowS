```python
import pandas as pd

df = pd.read_csv("data processing/Prolific, data collection - (Final, 4 ap) Dutch CrowS-Pairs (data verzameling) (3).csv")
df = df.sample(frac=1, random_state=42) 

df.head()

```

```python
#transform raw data from Prolific to usable,  flat data to split for validation
columns = list(df.columns)
sentence_pairs = []
number_of_sentences = 5

for index, row in df.iterrows():
    prolific_id = row["PROLIFIC_PID"]
    for i in range(1,6): 
        s1 = row[f"Zinspaar {i}a"]
        s2 = row[f"Zinspaar {i}b"]  
        group = row[f"Groep {i}"]
        label = row[f"Label {i}"]
        if pd.notna(s1) and pd.notna(s2):
            sentence_pairs.append({
                "id": len(sentence_pairs) + 1,
                "sentence1": s1,
                "sentence2": s2,
                "group": group,
                "label": label,
                "prolific_id": prolific_id
            }) 

df_flat = pd.DataFrame(sentence_pairs)
df_flat.to_csv('data processing/data_for_validation.csv', index=False)

assert len(df_flat) == len(df)*number_of_sentences
df_flat.head(10)
```

```python
# shuffle data and put in batches of 10
df = df_flat.sample(frac=1, random_state=42).reset_index(drop=True) 

batch_size = 10
batches = [df[i:i+batch_size] for i in range(0, len(df), batch_size)]

for i, batch in enumerate(batches):
    batch.to_csv(f"data processing/validation_batches/batch_{i+1}.csv", index=False)
```

```python
import json
import os
import pandas as pd

folder_path = "data processing/validation_batches"  # your CSV folder
output_json = "batches.json"

all_batches = []

for filename in sorted(os.listdir(folder_path)):
    if filename.startswith("batch_") and filename.endswith(".csv"):
        batch_id = os.path.splitext(filename)[0]
        df = pd.read_csv(os.path.join(folder_path, filename))

        # Turn each row into a full data pair
        pairs = []
        for _, row in df.iterrows():
            pairs.append({
                "id": int(row["id"]),
                "sentence1": row["sentence1"],
                "sentence2": row["sentence2"],
                "group": row["group"],
                "label": row["label"],
            })

        all_batches.append({
            "id": batch_id,
            "pairs": pairs
        })

# Save as JSON
with open(output_json, "w", encoding="utf-8") as f:
    json.dump({"batches": all_batches}, f, ensure_ascii=False, indent=2)

print(f"✅ Saved {len(all_batches)} batches to {output_json}")
```
